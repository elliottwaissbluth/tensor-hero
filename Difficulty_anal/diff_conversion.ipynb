## Creating the function that converts notes arrays from a higher difficulty to a lower one

The function will take two inputs:
-  An expert notes array
-  A transition probability matrix, $M$

Ultimately we will map the expert notes array to the lower difficulty by randomly sampling a note from the row in the transition probability matrix corresponding to the note present in the expert notes array.

For example, in a simplified 3-note world, the expert notes array might look like this:

\begin{bmatrix}
0 & 1 & 0 & 0 & 0 & 0 & 0 & 3 & 0 & 0 & 0
\end{bmatrix}

And $M$ probably looks something like this for expert to medium

\begin{bmatrix}
0.2 & 0.6 & 0.1 & 0.1 \\
0.3 & 0.1 & 0.5 & 0.1 \\
0.2 & 0.1 & 0.1 & 0.6 \\
\end{bmatrix}

Where each row, $i$, is encoded as
\begin{bmatrix}
P_{i,0} & P_{i,1} & P_{i,2} & P_{i,3}
\end{bmatrix}

So the first note in the notes array is a $1$, we would randomly sample from the first row in the transition matrix to determine what the one is translated to.

\begin{bmatrix}
P_{1,0} & P_{1,1} & P_{1,2} & P_{1,3}
\end{bmatrix}

\begin{bmatrix}
0.2 & 0.6 & 0.1 & 0.1
\end{bmatrix}

After randomly sampling, we find that a $2$ is selected. Now we will do a random sampling for the next note in the notes array, a 3.

\begin{bmatrix}
P_{3,0} & P_{3,1} & P_{3,2} & P_{3,3}
\end{bmatrix}

\begin{bmatrix}
0.2 & 0.1 & 0.1 & 0.6
\end{bmatrix}

After randomly sampling, we find $0$ is selected and the note is dropped, the new medium notes array looks like this.
\begin{bmatrix}
0 & 2 & 0 & 0 & 0 & 0 & 0 & 0 & 0 & 0 & 0
\end{bmatrix}
