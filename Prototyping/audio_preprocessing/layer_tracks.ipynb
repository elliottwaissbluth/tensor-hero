```python
from pathlib import Path
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import soundfile as sf
import sys

GH_dir = Path.cwd().parent.parent / 'Training_Data' / 'gold_standard_rock'

# Get's a list of ogg files from a directory containing .ogg
find_oggs = lambda file_list: [x for x in file_list if (x.endswith('.ogg') and x != 'preview.ogg' and x!= 'layered.ogg')]

non_22050_files = []
# Loop through all the songs, grabbing the component .ogg files
for song_dir in tqdm([GH_dir / x for x in os.listdir(GH_dir)]):
  print(f'song_dir: {song_dir}')
  ogg_files = [song_dir / x for x in find_oggs(os.listdir(song_dir))]
  print(f'num ogg files in song_dir: {len(ogg_files)}')
  print(f'ogg_files: {ogg_files}')
  # Load the songs in as a list of numpy arrays
  songs = []
  for x in ogg_files:
    song, sr = librosa.load(x)
    print(f'song.shape: {song.shape}')
    print(f'song.shape[0]: {song.shape[0]}')
    print(f'sr: {sr}')
    if sr != 22050:
      non_22050_files.append(song_dir)
    songs.append(song)
    print(f'after appending, songs length = {len(songs)}')
  print('out of the loop')
  print(f'number of loaded song arrays in songs: {len(songs)}')
  # Find the max len
  max_len = max([x.shape[0] for x in songs])
  print(f'max_len: {max_len}')
  final_song = np.zeros(max_len)

  # Layer the songs if they are equal to max_len
  print(f'layering {len(songs)} songs')
  for song in songs:
    print(f'song.shape[0]: {song.shape[0]}')
    if song.shape[0] == max_len:
      final_song += song
  
  # Write the song
  sf.write(str(song_dir / 'layered.ogg'), final_song, 22050, format='ogg')
```
