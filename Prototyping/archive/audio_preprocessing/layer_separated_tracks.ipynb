## The purpose of this notebook is to test the layering process for songs that come pre-separated
We are interested in layering these tracks so that we might use the data for our ML model. Since the model's input will be non-separated tracks, we want to combine these tracks into a single song.

```python
from pathlib import Path
import librosa
import numpy as np

# Let's load up some tracks
bass, bass_sr = librosa.load(str(Path.cwd() / 'layered_tracks' / 'Rag Doll' / 'bass.ogg'))
guitar, guitar_sr = librosa.load(str(Path.cwd() / 'layered_tracks' / 'Rag Doll' / 'guitar.ogg'))
preview, preview_sr = librosa.load(str(Path.cwd() / 'layered_tracks' / 'Rag Doll' / 'preview.ogg'))
rhythm, rhythm_sr = librosa.load(str(Path.cwd() / 'layered_tracks' / 'Rag Doll' / 'rhythm.ogg'))
song, song_sr = librosa.load(str(Path.cwd() / 'layered_tracks' / 'Rag Doll' / 'song.ogg'))
```

```python
print(f'bass len: {len(bass)}')
print(f'guitar len: {len(guitar)}')
print(f'preview len: {len(preview)}')
print(f'rhythm len: {len(rhythm)}')
print(f'song len: {len(song)}')
```

```python
import sounddevice as sd

sd.play(song+guitar+rhythm, song_sr)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(bass)
plt.title('bass')


plt.figure()
plt.plot(guitar)
plt.title('guitar')

plt.figure()
plt.plot(preview)
plt.title('preview')

plt.figure()
plt.plot(rhythm)
plt.title('rhythm')

plt.figure()
plt.plot(song)
plt.title('song')

plt.figure()
plt.plot(song+guitar+rhythm)
plt.title('song+guitar+rhythm')
```
