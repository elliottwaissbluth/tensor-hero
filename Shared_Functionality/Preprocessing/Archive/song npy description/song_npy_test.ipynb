```python
import numpy as np

spec = np.load('song.npy')
print(spec.shape)
```

## dim = [channel, frequency, time]

channel - Each channel (i.e spec[0], spec[1], spec[2]) contains an individual spectrogram. The difference between these spectrograms is the resolution as determined by the n_fft parameter

frequency - The frequency dimension is split into 81 dimensions. These are 81 log-mel bins (google mel-spectrogram for a better explanation!)

time - Each time slice represents 10ms of audio data, 70ms of empty data is prepended and appended to the song.

```python
import matplotlib.pyplot as plt
import librosa.display      # Have to import librosa.display specifically

# Let's display this spectrogram
for i in range(3):
    plt.figure() # librosa.display.specshow() is built off of matplotlib so this works
    librosa.display.specshow(spec[i,:,1000:1500], y_axis='mel', cmap='magma')
    plt.colorbar(format="%+2.f dB")
    
```
