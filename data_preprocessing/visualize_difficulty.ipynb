## Test whether the different difficulties are the same or not

```python
import os
import numpy as np
diff_dataset_path = Path.cwd().parent / 'Training_Data' / 'difficulty_dataset' / 'jackie_charts'

no_diff_song_paths = []
for song in [diff_dataset_path / x for x in os.listdir(diff_dataset_path)]:
    if '.DS_Store' in str(song):
        continue
    try:
        expert = np.load(song / 'Expert_simplified.npy')
        hard = np.load(song / 'Hard_simplified.npy')
        medium = np.load(song / 'Medium_simplified.npy')
        easy = np.load(song / 'Easy_simplified.npy')

        if list(easy) == list(hard) == list(medium) == list(easy):
            no_diff_song_paths.append(song)
    except:
        continue 
    
print(f'{len(no_diff_song_paths)} / {len(os.listdir(diff_dataset_path))} do not differentiate by difficulty') 
```
