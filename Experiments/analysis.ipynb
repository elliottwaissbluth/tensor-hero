```python
import pandas as pd

def label_freq_type_table(freq_type_table):
    '''
    Takes an unlabeled freq_type_table and applies labels to columns and rows
    for better interpretability
    '''
    row_labels = ['None', 'Single', 'Double', 'Triple', 'Quad', 'Five', 'Open']
    column_labels = ['Type', 'Truth', '% of total (Truth)','Observed', '% of total (Observed)']

    df = pd.DataFrame(columns = column_labels)
    df['Type'] = row_labels
    df['Truth'] = freq_type_table[:,0]
    df['% of total (Truth)'] = freq_type_table[:,1]
    df['Observed'] = freq_type_table[:,2]
    df['% of total (Observed)'] = freq_type_table[:,3]
    df.set_index('Type')
    return df

m1_pre_df = label_freq_type_table(m1_pre_sep['freq_type_table'])
m4_pre_df = label_freq_type_table(m4_pre_sep['freq_type_table'])
m4_post_df = label_freq_type_table(m4_post_sep['freq_type_table'])
```
